```python
import pandas as pd
```

```python
# setting to display all the columns and rows in the data
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# load both the datasets into dataframes
df_flats = pd.read_csv('flats_cleaned.csv')
df_houses = pd.read_csv('houses_cleaned.csv')
```

```python
# as the columns are same, concatenate both the dataframes
df = pd.concat([df_flats, df_houses], ignore_index=True)
df.shape
```

```python
# shuffle the rows of the concatenated df
df = df.sample(frac=1.0, random_state=42,ignore_index=True)
df.head(5)
```

```python
# flats vs houses
df.property_type.value_counts()
```

```python
# create a new column to store the sector information of the property
df.insert(loc=3, column='sector', value=df['property_name'].str.split('in').str.get(1).str.replace('Gurgaon', '').str.lower().str.strip())
df.sample(5)
```
