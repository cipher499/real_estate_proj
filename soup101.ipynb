```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df_final = pd.DataFrame()

for j in range(1, 11):

    url = "https://www.ambitionbox.com/list-of-companies?page={}".format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers).text

    soup = BeautifulSoup(webpage, 'lxml')

    companies = soup.find_all('div', class_='companyCardWrapper')

    name = []
    rating = []
    reviews = []
    ctype = []
    hq = []
    old = []
    num_emp = []

    for i in companies:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append('NA')
        
        try:
            rating.append(i.find('span', class_='companyCardWrapper__companyRatingValue').text.strip())
        except:
            rating.append('NA') 

        try:       
            reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
        except:
            reviews.append('NA')   

        try:             
            ctype.append(i.find('span', class_ = 'companyCardWrapper__interLinking').text.strip().split('|')[0])
        except:
            ctype.append('NA')

        try:
            num_emp.append(i.find('span', class_ = 'companyCardWrapper__interLinking').text.strip().split('|')[1])
        except:
            num_emp.append('NA')

        try:
            hq.append(i.find('span', class_ = 'companyCardWrapper__interLinking').text.strip().split('|')[4])
        except:
            hq.append('NA')
            
        try:
            old.append(i.find('span', class_ = 'companyCardWrapper__interLinking').text.strip().split('|')[3])
        except:
            old.append('NA')
        
    df = pd.DataFrame({"name": name, "rating": rating, "reviews": reviews, "ctype": ctype, "old": old, "hq": hq, "num_emp": num_emp})

    df_final = df_final.append(df, ignore_index=True)
```
